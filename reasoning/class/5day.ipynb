```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('./data/sales1.csv', encoding='cp949')
df2 = pd.read_csv('./data/sales2.csv', encoding='cp949')
df3 = pd.read_csv('./data/sales3.csv', encoding='cp949')
```

```python
df12_left  =  pd.merge(df1, df2, how='left', on='ORDERID')
df12_left
```

```python
df = pd.merge(df12_left, df3, how='left', on='ORDERID')
df
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%m/%d/%Y')
df.head()
```

```python
from datetime import datetime
xmas = datetime.strptime('2024-12-25', '%Y-%m-%d').date()
xmas
```

```python
df['TO_XMAS'] = (xmas - df['DATE'].dt.date).astype(str).str[:5].astype(int)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/sales.csv', encoding='cp949')
```

```python
plt.hist(df['TOTAL_PRICE'])
plt.show()
```

```python
plt.hist(df['TOTAL_PRICE'], bins=20, color='green')
plt.show()
```

```python
man_price = df[df['GENDER']=='Male']['TOTAL_PRICE']
```

```python
woman_price = df[df['GENDER']=='Female']['TOTAL_PRICE']
```

```python
plt.hist(man_price, bins=20, color='cornflowerblue', alpha=0.4, edgecolor='black', linewidth=1, label='man')
plt.hist(woman_price, bins=20, color='coral', alpha=0.4, edgecolor='black', linewidth=1, label='woman')
plt.xlabel('Total price')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
category_counts = df['CITY'].value_counts()
print(category_counts)
```

```python
city_idx = category_counts.index
print(city_idx)
```

```python
colors = ['red', 'blue', 'orange']
plt.bar(x=city_idx, height=category_counts, color=colors)
plt.show()
```

```python
city_gender = pd.crosstab(df['CITY'], df['GENDER'])
city_gender
```

```python
city_gender.index
```

```python
city_gender.columns
```

```python
plt.bar(x=city_gender.index, height=city_gender['Male'], color='lightcoral', label='Male')
plt.bar(x=city_gender.index, height=city_gender['Female'], color='paleturquoise', label='Female', bottom=city_gender['Male'])
plt.ylim(0, 500)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
city_price = df.groupby('CITY')['TOTAL_PRICE'].mean()
city_price
```

```python
plt.bar(x=city_gender.index, height=city_gender['Male'],  color='lightcoral', label='Male')
plt.bar(x=city_gender.index, height=city_gender['Female'], color='paleturquoise', label='Female', bottom=city_gender['Male'])
plt.ylim(0, 500)
plt.plot(city_gender.index, city_price, '^-')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
bar1 = ax.bar(x=city_gender.index, height=city_gender['Male'],  color='lightcoral', label='Male')
bar2 = ax.bar(x=city_gender.index, height=city_gender['Female'], color='paleturquoise', label='Female', bottom=city_gender['Male'])
ax.bar_label(bar1, label_type='center')
ax.bar_label(bar2, label_type='center')
plt.ylim(0, 500)
plt.plot(city_gender.index, city_price, '^-')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
