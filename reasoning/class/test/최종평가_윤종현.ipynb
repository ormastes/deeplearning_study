````python
# 5. 위 모형들에 대해 그리드 서치를 사용하여 각 모형의 하이퍼파라미터를 최적화 (4점)
#Decision Tree 하이퍼 파라미터 후보: 'max_depth': [3, 5, 7, 10], 'criterion': ['gini', 'entropy', 'log_loss']
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
DecisionTreeClassifierf1 = None
DecisionTreeClassifierrecall = None
DecisionTreeClassifierpre = None
DecisionTreeClassifieracc = 1000000
for max_depth in [3, 5, 7, 10]:
    for criterion in ['gini', 'entropy', 'log_loss']:
        clf_dt = DecisionTreeClassifier(max_depth=max_depth,criterion=criterion, random_state=0).fit(X_tn, y_tn)
        pred = clf_dt.predict(X_te)
        _acc = accuracy_score(y_te, pred)
        if _acc < DecisionTreeClassifieracc:
            DecisionTreeClassifieracc = _acc
            DecisionTreeClassifierpre = precision_score(y_te, pred, average='macro', zero_division=1)
            DecisionTreeClassifierf1 = f1_score(y_te, pred, average='macro')
            DecisionTreeClassifierrecall = recall_score(y_te, pred, average='macro')
            _max_depth = max_depth
            _criterion = criterion
print('DecisionTreeClassifier max_depth', _max_depth, 'criterion', _criterion)
        
````

````python
# 5. 위 모형들에 대해 그리드 서치를 사용하여 각 모형의 하이퍼파라미터를 최적화 (4점)
#Support Vector Machine 하이퍼 파라미터 후보: 'C': [0.1, 1, 10], 'gamma': [0.01, 0.1, 1, 2]
SVCpre = None
SVCacc = 1000000
for C in [0.1, 1, 10]:
    for gamma in [0.01, 0.1, 1, 2]:
        clf_dt = SVC(C=C,gamma=gamma, random_state=0).fit(X_tn, y_tn)
        pred = clf_dt.predict(X_te)
        _acc = accuracy_score(y_te, pred)
        if _acc < SVCacc:
            SVCacc = _acc
            SVCpre = precision_score(y_te, pred, average='macro', zero_division=1)
            SVCf1 = f1_score(y_te, pred, average='macro')
            SVCrecall = recall_score(y_te, pred, average='macro')
            _C = C
            _gamma = gamma
print('SVC C', _C , 'gamma', _gamma)
        
````

````python
# 5. 위 모형들에 대해 그리드 서치를 사용하여 각 모형의 하이퍼파라미터를 최적화 (4점)
#K Nearest Neighbor 하이퍼 파라미터 후보: 'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']
KNeighborsClassifierpre = None
KNeighborsClassifieracc = 1000000
for n_neighbors in [3, 5, 7]:
    for weights in ['uniform', 'distance']:
        clf_dt = KNeighborsClassifier(n_neighbors=n_neighbors,weights=weights).fit(X_tn, y_tn)
        pred = clf_dt.predict(X_te)
        _acc = accuracy_score(y_te, pred)
        if _acc < KNeighborsClassifieracc:
            KNeighborsClassifieracc = _acc
            KNeighborsClassifierpre = precision_score(y_te, pred, average='macro', zero_division=1)
            KNeighborsClassifierf1 = f1_score(y_te, pred, average='macro')
            KNeighborsClassifierrecall = recall_score(y_te, pred, average='macro')
            _weights = weights
            _n_neighbors =n_neighbors
print('KNeighborsClassifier weights', _weights , 'n_neighbors', _n_neighbors)
````

````python
# 6. 각 최적화 모형별 `accuracy`, `recall`, `precision`, `f1-score`값을 출력 (2점)
print('DecisionTreeClassifier accuracy', DecisionTreeClassifieracc)
print('SVC accuracy', SVCacc)
print('KNeighborsClassifier accuracy', KNeighborsClassifieracc)        
        
````

### 9. 딥러닝 (10점)

1. 다음 이미지를 참고하여 파이토치로 딥러닝 신경망 모델을 구현하는 class를 완성하시오.(10점)<br>

````python
import torch.nn as nn

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        INPUT = 12
        NODE_1 = 20
        NODE_2 = 15
        NODE_3 = 20
        NODE_4 = 10
        NODE_5 = 5
        model = nn.Sequential(
            nn.BatchNorm1d(),
            nn.ReLU(),
            nn.Linear(INPUT, NODE_1),
            nn.Dropout(p=4/NODE_1),
            nn.BatchNorm1d(),
            nn.ReLU(),
            nn.Linear(NODE_1, NODE_2),
            nn.Dropout(p=3/NODE_2),
            nn.BatchNorm1d(),
            nn.ReLU(),
            nn.Linear(NODE_2, NODE_3),
            nn.Dropout(p=2/NODE_3),
            nn.BatchNorm1d(),
            nn.ReLU(),
            nn.Linear(NODE_3, NODE_4),
            nn.Dropout(p=1/NODE_4),
            nn.BatchNorm1d(),
            nn.Linear(NODE_4, NODE_5))


    def forward(self, x):
        x = self.model(x)
        
        return x
````
