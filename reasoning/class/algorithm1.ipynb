```python
import pandas as pd
df = pd.read_csv("./data/wine_data.csv")
df
```

```python
features = ['Alcohol','Malic','Ash','Alcalinity',	'Magesium',	'Phenols',	'Flavanoids',	'Nonflavanoids',	'Proanthocyanins',	'Color',	'Hue',	'Dilution'
            , 'Proline']
X = df[features]
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
from sklearn.linear_model import LogisticRegression
clf_logi_l2 = LogisticRegression(penalty='l2')
clf_logi_l2.fit(X_tn_std, y_tn)
```

```python
clf_logi_l2.coef_
```

```python
clf_logi_l2.intercept_
```

```python
pred_logicstic = clf_logi_l2.predict(X_te_std)
pred_logicstic
```

```python
prd_prob = clf_logi_l2.predict_proba(X_te_std)
prd_prob
```

```python
from sklearn.naive_bayes import GaussianNB
clf_gnb = GaussianNB()
clf_gnb.fit(X_tn_std, y_tn)
pred_gnb = clf_gnb.predict(X_te_std)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_te, pred_gnb)
conf_matrix
```

```python
from sklearn.metrics import classification_report
class_report = classification_report(y_te, pred_gnb)
print(class_report)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf_tree = DecisionTreeRegressor(random_state=0)
clf_tree.fit(X_tn_std, y_tn)
```

```python
pred_tree = clf_tree.predict(X_te_std)
pred_tree
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_te, pred_tree)
accuracy
```

```python
from sklearn import svm
clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(X_tn_std, y_tn)
```

```python
pred_svm = clf_svm.predict(X_te_std)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_te, pred_svm)
conf_matrix
```
