```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

```

```python
df = pd.read_csv("./data/iris.csv")
```

```python
df
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
X = df[features]
y = df['class']
X
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, Y_tn, Y_te = train_test_split(X, y, random_state=0)
X_tn
```

```python
X_te.head()
```

```python
Y_tn.head()
```

```python
Y_te.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
X_tn_std
```

```python
from sklearn import svm
clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(X_tn_std, Y_tn)
```

```python
pred_svm = clf_svm.predict(X_te_std)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(Y_te, pred_svm)
conf_matrix
```

```python
from sklearn.svm import SVR
clf_svr = SVR(kernel='linear')
clf_svr.fit(X_tn_std, Y_tn)
```

```python
svr_pred = clf_svr.predict(X_te_std)
svr_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_te, svr_pred)
```

```python
mse
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import VotingRegressor
```

```python
clf1 = LinearRegression()
clf2 = SVR(kernel='linear')
clf3 = DecisionTreeRegressor()

clf_vot = VotingRegressor(estimators=[
    ('lr', clf1),
    ('svm', clf2),
    ('tree', clf3)
],weights=[1,1,1])
clf_vot.fit(X_tn_std, Y_tn)
```

```python
vote_pred = clf_vot.predict(X_te_std)
vote_pred
```

```python
mean_squared_error(Y_te, vote_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf.fit(X_tn_std, Y_tn)
```

```python
pred_rf = clf_rf.predict(X_te_std)
pred_rf
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_te, pred_rf)
```

```python
import pandas as pd
df = pd.read_csv("./data/house_prices.csv")
df
```

```python
df.columns
```

```python
features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
X = df[features]
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)

from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf_rf = RandomForestRegressor(max_depth=2, random_state=0)
clf_rf.fit(X_tn_std, y_tn)
```

```python
pred_rf = clf_rf.predict(X_te_std)
```

```python
mean_squared_error(y_te, pred_rf)
```

```python
from sklearn.ensemble import AdaBoostRegressor
clf_ada = AdaBoostRegressor(random_state=0)
clf_ada.fit(X_tn_std, y_tn)
```

```python
ada_pred = clf_ada.predict(X_te_std)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_te, ada_pred)
```
